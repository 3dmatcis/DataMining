```python
import pandas as pd
```

```python
empresas = pd.read_html('site.html',encoding='utf-8')
empresas_df = pd.DataFrame(columns='Nome Site Setores Produtos'.split(' '))
```

```python
nomes = []
sites = []
setores = []
produtos = []

for empresa in empresas:
    nome = empresa[0][0]
    site = empresa[0][1][6:]
    setor = empresa[0][2][10:]  
    try:
        produto = empresa[0][3][10:]
        
    except:
        pass
    nomes.append(nome)
    sites.append(site)
    setores.append(setor)
    produtos.append(produto)

empresas_df['Nome'] = nomes
empresas_df['Site'] = sites
empresas_df['Setores'] = setores
empresas_df['Produtos'] = produtos
empresas_df
```

```python
empresas_df.to_excel('planilhas/ABIMO-implantes.xls',index=False)
```
